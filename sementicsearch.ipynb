```python
from rapidfuzz import fuzz, utils
def _best_fuzzy(a: str, b: str) -> float:
    proc = utils.default_process
    scores = [
        fuzz.QRatio(a, b, processor=proc),
        fuzz.token_set_ratio(a, b, processor=proc),
        fuzz.partial_ratio(a, b, processor=proc),
    ]
    return round(float(max(scores)), 2)
```
